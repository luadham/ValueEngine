```python
from engine import Value
```

```python
X_ds = [1, 2, 3, 4]
```

```python
y_ds = [1, 2, 3, 4]
```

```python
w = Value(0.2);
b = Value(0);
```

```python
# Train Model

epochs = 100
for i in range(epochs):
    y_pred = [w * x + b for x in X_ds]
    loss = sum( (y_p - y_true)**2 for y_true, y_p in zip(y_ds, y_pred) ) / len(y_ds)
    print(loss.data)
    w.grad = 0
    b.grad = 0
    loss.backward()
    w.data += -(0.1 * w.grad)
    b.data += -(0.1 * b.grad)
```
